Список потраченных денег за последние 4 месяца

Количество затрат:
Февраль – 10203 м.

Март – 8868 м.

Апрель – 5642 м.

Май - 4466 м


```python
import matplotlib.pyplot as plt
%matplotlib inline
fruits = ["Февраль","Март", "Апрель", "Май"]
counts = [10203,8849,5642, 4466]
plt.bar(fruits, counts)
plt.title("Количество затрат")
plt.xlabel("Название месяцев")
plt.ylabel("Сумма")
```
